```python
import pandas as pd
import numpy as np
```

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
for i in range(1,5):
    url = 'https://www.imdb.com/search/title/?count=100&groups=top_1000&sort=user_rating'+str(i)
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
movie_name = []
year = []
time = []
rating = []
metascore = []
votes = []
gross = []
```

```python
movie_data = soup.findAll('div', attrs= {'class':'lister-item mode-advanced'})
```

```python
for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
    
    year_of_relese = store.h3.find('span', class_='lister-item-year text-muted unbold').text.replace('(','').replace(')','')
    year.append(year_of_relese)
    
    runtime = store.p.find('span', class_='runtime').text.replace(' min','')
    time.append(runtime)
    
    rate = store.find('div', class_='inline-block ratings-imdb-rating').text.replace('\n','')
    rating.append(rate)
    
    meta = store.find('span', class_='metascore').text.replace('        ','') if store.find('span', class_='metascore') else '^^^^'
    metascore.append(meta)
    
    value = store.find_all('span', attrs = {'name': 'nv'})
    
    vote = value[0].text
    votes.append(vote)
    
    grosses = value[1].text if len(value) >1 else '*****'
    gross.append(grosses)
```

```python
movie_Df = pd.DataFrame({'Name of the movie':movie_name, 'Year of relesing':year, 'Movie length':time, 'Rating of the movie':rating, 'Metascore':metascore, 'Votes':votes, 'Gross':gross})
```
