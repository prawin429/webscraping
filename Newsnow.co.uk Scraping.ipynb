```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url = 'https://www.newsnow.co.uk/h/Industry+Sectors/Food+&+Drink/Beverages/The+Coca-Cola+Company'
```

```python
res = requests.get(url)
```

```python
res.status_code
```

```python
soup = BeautifulSoup(res.content, 'lxml')
```

```python
news = soup.find('div', class_ = 'hl__inner')
```

```python
news.text
```

```python
news1 = soup.find_all('div', class_ = 'hl__inner')
```

```python
news1
```

```python
news1 = [news.text for news in news1]
```

```python
news1
```

```python
news1[0]
```

```python
import pandas as pd
```

```python
dataset = pd.DataFrame()
```

```python
dataset['News'] = news1
```

```python
dataset
```

```python
dataset.to_excel('News.xlsx')
```
